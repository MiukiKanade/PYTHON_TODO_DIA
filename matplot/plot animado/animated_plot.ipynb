```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation
import matplotlib as mpl 
mpl.rcParams['animation.ffmpeg_path'] = r'C:\\Users\\erikj\\Downloads\\ffmpeg\\bin\\ffmpeg.exe'

fig = plt.figure()
ax = plt.axes(xlim=(0, 2), ylim=(-2, 2))
line, = ax.plot([], [], lw=2)

def init():
    line.set_data([], [])
    return line,

def animate(frames):
    #frames Ã© quantidade de frames. Funciona de modo iterativo
    x = np.linspace(0, 2, 1000) #eixo x
    y = np.sin(2 * np.pi * (x - 0.01 * frames)) #eixo y
    line.set_data(x, y)
    return line,

anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=200, interval=20, blit=True)
#salvar como gif
f = r"animation.gif" 
writergif = animation.PillowWriter(fps=30) 
anim.save(f, writer=writergif)

#salvar como avi
'''f = r"c://Users/erikj/Desktop/animation.avi" 
writervideo = animation.FFMpegWriter(fps=60)
anim.save(f, writer=writervideo)'''

plt.show()
#https://jakevdp.github.io/blog/2012/08/18/matplotlib-animation-tutorial/
```
